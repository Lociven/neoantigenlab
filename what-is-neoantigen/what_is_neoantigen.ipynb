# What Is a Neoantigen?
**Companion notebook for [NeoantigenLab Post #1](https://www.neoantigenlab.com)**

This notebook illustrates the core concepts from the blog post with simple code examples:
- What makes a mutation a neoantigen candidate
- The MHC binding funnel
- Visualizing mutation types and their immunogenic potential

```python
# Neoantigen candidate funnel — conceptual illustration
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

stages = [
    ('Somatic mutations', 200),
    ('Coding / nonsynonymous', 80),
    ('Strong MHC binders\n(IC50 < 500 nM)', 25),
    ('Predicted presented\n(processing score)', 12),
    ('Immunogenic\n(T cell response)', 5),
]

fig, ax = plt.subplots(figsize=(8, 5))
colors = ['#2c7bb6', '#1a9bbf', '#0d7a8a', '#085f6e', '#053d47']

for i, (label, count) in enumerate(stages):
    ax.barh(i, count, color=colors[i], alpha=0.85)
    ax.text(count + 2, i, f'{count}', va='center', fontsize=11)

ax.set_yticks(range(len(stages)))
ax.set_yticklabels([s[0] for s in stages], fontsize=10)
ax.set_xlabel('Number of candidates')
ax.set_title('Neoantigen Identification Funnel (typical solid tumor)', fontsize=13)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('neoantigen_funnel.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Mutation types and neoantigen potential

| Mutation type | Novel peptide? | Typical immunogenicity |
|--------------|---------------|------------------------|
| SNV (missense) | 1 amino acid change | Moderate |
| Frameshift indel | Entire downstream sequence | High |
| In-frame indel | 1+ amino acids | Moderate |
| Gene fusion | Junction peptide | Variable |
| Synonymous SNV | None | None |

## Key concepts recap

1. **Somatic mutation** → altered protein → novel peptide fragment
2. **Proteasomal cleavage** → 8–11 aa peptide
3. **TAP transport** → ER
4. **HLA binding** → peptide-MHC complex on cell surface
5. **TCR recognition** → T cell activation

Each step is a filter. Most mutations don't survive all five.
